```python
import pandas as pd
import io
from google.colab import files

print("Please upload the 'phishing website dataset' file when prompted.")
uploaded = files.upload()

if not uploaded:
    print("No file was uploaded. Please try again and select your 'phishing website dataset.csv' file.")
else:
    file_path = None
    for fn in uploaded.keys():
        print('User uploaded file "{name}" with length {length} bytes'.format(
            name=fn, length=len(uploaded[fn])))
        file_path = fn

    if file_path:
        try:
            df = pd.read_csv(io.StringIO(uploaded[file_path].decode('utf-8')))

            print(f"'{file_path}' loaded successfully.")
            print("Displaying the first 5 rows of the DataFrame:")
            print(df.head())
        except Exception as e:
            print(f"An error occurred while loading the dataset: {e}")
    else:
        print("Error: No valid file path was determined after upload.")
```

```python
print("Displaying the first 5 rows of the DataFrame:")
print(df.head())

print("\nChecking for missing values:")
print(df.isnull().sum())

print("\nGenerating descriptive statistics for numerical columns:")
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

result_counts = df['Result'].value_counts()

plt.figure(figsize=(7, 5))
sns.countplot(x='Result', data=df, palette='viridis')


plt.title('Distribution of Phishing vs. Legitimate Websites')
plt.xlabel('Website Type (-1: Phishing, 1: Legitimate)')
plt.ylabel('Number of Websites')

plt.xticks(ticks=[0, 1], labels=['Phishing', 'Legitimate'])


for index, value in enumerate(result_counts):
    plt.text(index, value + 1000, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


result_counts = df['Result'].value_counts()


plt.figure(figsize=(7, 5))
sns.countplot(x='Result', data=df, hue='Result', palette='viridis', legend=False)


plt.title('Distribution of Phishing vs. Legitimate Websites')
plt.xlabel('Website Type (-1: Phishing, 1: Legitimate)')
plt.ylabel('Number of Websites')

plt.xticks(ticks=[0, 1], labels=['Phishing', 'Legitimate']) # Assuming -1 and 1 are mapped to 0 and 1 for plotting


for index, value in enumerate(result_counts):
    plt.text(index, value + 1000, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
